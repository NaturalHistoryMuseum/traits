```python
import spacy
import re
from spacy import displacy
import pandas as pd
import numpy as np


from adept.config import RAW_DATA_DIR
```

```python
excel_file = RAW_DATA_DIR / 'angiosperm.xlsx'

sheets = ['ecoflora', 'wikipedia', 'flora_of_china']

dfs = [pd.read_excel(excel_file, sheet_name=sheet) for sheet in sheets]

df = pd.concat(dfs)
```

```python
df.shape
```

```python
def aggregate(cell):

    _ensure_set = lambda x: x if isinstance(x, set) else set([x])

    if cell.dtype in ['int64', 'float64']:
        return round(cell.mean(), 2)
    else:        
        if combined := [_ensure_set(c) for c in cell if not pd.isnull(c)]:
            return set.union(*combined)
        
    return None
    
df.groupby('taxon').aggregate(aggregate)
```
