```python
from pathlib import Path
from spacy import displacy
import spacy
import joblib

from leventis.taxon_entity_matcher import TaxonEntityMatcher 
from leventis.bhl_page import BHLPage
from leventis.normalise import text_preprocessor
```

```python
MODEL_PATH = Path('../data/models')
```

```python
model = joblib.load(MODEL_PATH / 'naive-bayes-model.pkl') 
```

```python
nlp = spacy.load('en')

entity_matcher = TaxonEntityMatcher(model)

nlp.add_pipe(entity_matcher, after="ner") 
```

```python
bhl_page = BHLPage(52683973)

text = text_preprocessor(bhl_page.get_text())

doc = nlp(text)
```

```python
colors = {"TAXON": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}
options = {"colors": colors}
displacy.render(doc, style="ent", options=options)
```
