```python
%load_ext autoreload
%autoreload 2

from spacy import displacy

import pandas as pd
import mysql.connector as sql
```

```python
from leventis.parse_page import ParsePage
```

```python
db_connection = sql.connect(database='pup', user='root')

query = '''
SELECT bc.page_id, pt.trait_term, pn.pup_name AS taxon 
FROM bhl_citations bc 
    INNER JOIN pup_names_sp pn ON pn.pup_name_id = bc.pup_id 
    INNER JOIN trait_page tp ON tp.page_id = bc.page_id 
    INNER JOIN x_pup_traits pt ON tp.trait_id=pt.trait_id 
WHERE pn.pup_higher_group = 'Dicot' AND bc.item_language = 'English';
'''

df = pd.read_sql(query, con=db_connection)

# df = df.groupby('page_id')['taxon'].agg(set).reset_index()

df.head()
```

```python
df = df[df['taxon'] == 'Morisonia americana']

df.head()
```

```python
results = {}


for page_id, taxa in df.itertuples(index=False):
    
    parsed_page = ParsePage(page_id)

    for taxon in taxa:
        
        results.setdefault(taxon, {})
        
        try:
            results[taxon][page_id] = parsed_page.traits[taxon]
        except KeyError:
            
            if taxon not in parsed_page.taxa:
                print(f"Taxon {taxon} not found in {page_id}")
            else:
                print(f"No traits found for {taxon} in {page_id}")
                
    print(results)
            
            
            

        
#         try:
            
        
        
        
        
#         print(taxon)
        
#     x = list(parsed_page.traits.values())
#     print(x)
    
#     s = x[0]
    
#     print(type(s))
    
#     e1 = next(iter(s))
#     e2 = next(iter(s))
#     e3 = next(iter(s))
    
#     if e1 == e2:
#         print('Y')
    
#     for l in x:
#         print(l)
    
    
#     print(l[0])
#     print(l[1])
    
#     if l[0] != l[2]:
#         print(f'--{l[0]}--')
#         print(f'--{l[2]}--')
#         print('N')
    
#     for i in x[0]:
#         print(i)
#         print(repr(i))
# #     print(type(x[0]))

# #     print(set(list(x[0])))
#     break
#     print(parsed_page.taxa)
    
#     traits = traits.get_traits(page_id)
    
#     break
    
#     taxa = set(taxa)
#     print(taxa)
    

# traits = Traits()

# for row in df.iterrows():
#     print(row['page_id'])
# #     traits = traits.get_traits(row['page_id'])
# #     print(traits)
    
#     break
```
