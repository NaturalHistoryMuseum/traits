```python
from pathlib import Path
from spacy import displacy
import spacy

from leventis.bhl_page import BHLPage
from leventis.normalise import text_preprocessor
```

```python
MODEL_PATH = Path('../data/models')
```

```python
nlp = spacy.load(MODEL_PATH / "en_sci_g4_md")
```

```python
bhl_page = BHLPage(52683973)

text = text_preprocessor(bhl_page.get_text())

doc = nlp(text)
```

```python
colors = {"T204": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"}
options = {"colors": colors}
displacy.render(doc, style="ent", options=options)
```
